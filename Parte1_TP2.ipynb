#Visualizaciones

## Bar plot

```python
hongos_por_color = df.groupby(['cap-color', 'class']).size().reset_index(name='count')
hongos_por_color
```

```python
import seaborn as sns

prueba_colores = sns.color_palette("bright")
colores = [prueba_colores[-1], "green"] #Colores=[comestible, venenoso]
print(prueba_colores)
prueba_colores
```

```python
import numpy as np
from matplotlib import pyplot as plt

# Cuento la cantidad de hongos venenosos y no venenosos para cada color
hongos_por_color = df[["class", "cap-color"]]
hongos_por_color["cap-color"] = df["cap-color"].map({"n":"brown", "b":"buff", "g":"gray", "r":"green", "p":"pink",\
"u":"purple", "e":"red", "w":"white", "y":"yellow", "l":"blue", "o":"orange", "k":"black"})
hongos_por_color = hongos_por_color.groupby(['cap-color', 'class']).size().reset_index(name='count')

# Creo el gráfico de barras
sns.barplot(data=hongos_por_color, x="cap-color", y="count", hue="class", palette=colores)  #stacked=True
plt.xlabel('Colores')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)
plt.title('Cantidad de hongos venenosos y no venenosos por color del sombrero')
plt.grid(axis="y")
plt.gca().set_axisbelow(True)

plt.show()
```

Se puede ver que la mayoria de hongos, tanto venenosos como comestibles tienen un color de sombrero marrón y que el color del sombrero no suele ser un factor muy determinante a excepción de el rojo o el naranja por ejemplo, que muestran que en su mayoría suelen ser venenosos.

## Violin plot

```python
#plt.figure(dpi=150)
sns.violinplot(data=df, x="class", y="cap-diameter", palette=colores[::-1])
plt.title("Clase de hongo según el diámetro del sombrero")
plt.ylim((-10,40))
lista = []
for i in range(-10,50,5):
  lista.append(i)
plt.yticks(lista)
plt.ylabel("Diametro del sombrero (cm)")
plt.xlabel("Tipo")
plt.grid(axis="y")
plt.gca().set_axisbelow(True)
plt.show()
```

Se puede notar que si el hongo tiene un sombrero con diámetro mayor a 35 cm es muy probable que sea comestible, en caso contrario, puede variar.

## Box plot

```python
#plt.figure(dpi=150)
sns.boxplot(data=df, x="class", y="stem-height", palette=colores[::-1])
plt.title("Clase de hongo según la altura del tallo")
#plt.ylim((-10,40))
lista = []
for i in range(0,31,5):
  lista.append(i)
plt.yticks(lista)
plt.grid(axis="y")
plt.ylabel("Altura del tallo (cm)")
plt.xlabel("Tipo")
plt.gca().set_axisbelow(True)
plt.show()
```

Se puede ver que si la altura del tallo es mayor a 25 cm el hongo suele ser comestible.

## Heatmap

```python
paleta2 = sns.color_palette("YlOrBr", as_cmap=True)
paleta = sns.cubehelix_palette(start=2, rot=0, dark=0, light=.95, as_cmap=True)
paleta
```

```python
df_heatmap_aux = df[["class", "season", "habitat"]]
df_heatmap_aux["season"] = df_heatmap_aux["season"].map({"w":"winter", "u":"summer", "a": "autumn", "s":"spring"})
df_heatmap_aux["class"] = df_heatmap_aux["class"].map({"Venenoso":1, "Comestible":0})
df_heatmap_aux["habitat"] = df_heatmap_aux["habitat"].map({"g":"grasses", "l":"leaves", "m":"meadows", "p":"paths", "h":"heaths", "u":"urban", "w":"waste", "d":"woods"})

df_heatmap = df_heatmap_aux.pivot_table(values="class", index="habitat", columns="season", aggfunc="sum")

nan_mask = df_heatmap.isnull()

df_heatmap
```

```python
plt.figure()
plt.title("Cantidad de hongos venenosos por habitat por temporada")
g = sns.heatmap(data=df_heatmap, square=True, cmap=paleta, mask=nan_mask)
g.set_facecolor('yellow')
plt.show()
```

Sin embargo, este Heatmap puede ser engañoso, porque si bien en el habitat "woods" en la temporada "autumn" (otoño) pareciera que hay muchos hongos venenosos, se puede ver que la cantidad total de hongos en ese hábitat y en esa temporada es significativamente mayor a los demás. Para los valores NaNs elegí poner un color amarillo para distinguirlos.

```python
hongos_por_habitat_por_temporada = df_heatmap_aux.groupby(["habitat", "season"]).size()
hongos_por_habitat_por_temporada
```

Por lo expuesto anteriormente, ahora paso a trabajar con porcentajes para que los datos tengan más sentido.

```python
df_aux = df.copy()
df_aux["season"] = df["season"].map({"w":"winter", "u":"summer", "a": "autumn", "s":"spring"})
df_aux["habitat"] = df_aux["habitat"].map({"g":"grasses", "l":"leaves", "m":"meadows", "p":"paths", "h":"heaths", "u":"urban", "w":"waste", "d":"woods"})

df_venenosos = df_aux[df_aux["class"] == "Venenoso"]
df_comestibles = df_aux[df_aux["class"] == "Comestible"]

df_venenosos_agrupados = df_venenosos.groupby(["habitat", "season", "class"]).size().reset_index(name='cantidad_venenosos')
df_comestibles_agrupados = df_comestibles.groupby(["habitat", "season", "class"]).size().reset_index(name='cantidad_comestibles')

df_mergeados = df_comestibles_agrupados.merge(df_venenosos_agrupados, on=["habitat", "season"], how="outer")
#df_mergeados

df_porcentajes = df_mergeados.copy()
df_porcentajes["porcentaje"] = df_mergeados["cantidad_venenosos"] / (df_mergeados["cantidad_comestibles"]+df_mergeados["cantidad_venenosos"]) * 100
df_porcentajes = df_porcentajes.fillna(0)
df_porcentajes
```

```python
df_heatmap2 = df_porcentajes.pivot_table(values="porcentaje", index="habitat", columns="season")
nan_mask2 = df_heatmap2.isnull()
df_heatmap2
```

Decido no rellenar los NaNs con 0 porque equivaldría a decir que el porcentaje que los hongos sean venenosos en determinado habitat y estación es 0 y eso es incorrecto.

```python
#plt.figure(dpi=150)
plt.title("Porcentaje de hongos venenosos por habitat por temporada")
g = sns.heatmap(data=df_heatmap2, square=True, cmap=paleta)
g.set_facecolor('yellow')
plt.show()
```

## Gráfico 5- Density plot

```python
df_g5 = df[["class", "stem-width"]]
df_g5
```

```python
#plt.figure(dpi=150)
sns.kdeplot(data=df_g5, x="stem-width", hue="class", palette=colores[::-1])
plt.title("Probabilidad del ancho del tallo")
lista = []
for i in range(-10,101,5):
  lista.append(i)
plt.xticks(lista)
plt.xlim((0,75))
plt.xlabel("Ancho del tallo (mm)")
plt.ylabel("Densidad")
plt.grid()
plt.gca().set_axisbelow(True)
plt.show()
```

Con este último gráfico se puede observar por ejemplo, que si un hongo tiene un tallo con ancho menor a 10mm, es más probable que sea venenoso a que sea comestible.
