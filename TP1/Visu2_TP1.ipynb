```python
#Arreglo el indice de las columnas
estaciones_max = est_max.set_axis(est_max.iloc[0], axis=1)
estaciones_max = estaciones_max[1:]
estaciones_max = estaciones_max.dropna(subset=['San Luis Tucuman'])

estacion_SL_max = estaciones_max[["Año hid", "San Luis Tucuman"]]
estacion_SL_max = estacion_SL_max.astype(float)
estacion_SL_max
```

```python
#Arreglo el indice de las columnas
estaciones_anuales = est_anuales.set_axis(est_anuales.iloc[0], axis=1)
# Elimina la primera fila, que ahora es redundante
estaciones_anuales = estaciones_anuales[1:]
estaciones_anuales = estaciones_anuales.dropna(subset=['San Luis Tucuman'])
estaciones_anuales
```

```python
bins = range(0, 301, 10)  # 0 a 200 de a 10 (cant de barras que quiero)
plt.figure(figsize=(11.0, 4.8))
# Crea el histograma utilizando Seaborn
palette = sns.color_palette("husl", n_colors=30)
sns.histplot(data=estacion_SL_max, x="San Luis Tucuman", bins = bins, color = "#6FDBEC") #Color celeste asemejando al agua

# Agrega los títulos y etiquetas de los ejes
plt.title('LLuvia más fuerte de cada año por rango de precipitación')
plt.xlabel('Precipitación de la lluvia más fuerte del año(mm)')
plt.xticks(bins) #pongo la misma cantidad de etiquetas que de barras
plt.xticks(rotation=45)
plt.ylabel('Cantidades de lluvias que\nalcanzaron la precipitación')
plt.show()
```
