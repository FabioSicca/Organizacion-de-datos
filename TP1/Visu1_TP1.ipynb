```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
datos_soja = pd.read_csv('/content/drive/MyDrive/Organizacion de datos/VisusOrga/datasetRofex.csv')
```

```python
datos_soja_filtered = datos_soja[["Cierre", "Fecha","DolarB.N."]]

#Reemplazo las comas por puntos en la col DolarB.N. para poder interpretarlo como float luego
datos_soja_filtered["DolarB.N."] = datos_soja_filtered["DolarB.N."].replace(',', '.', regex=True)
datos_soja_filtered["DolarB.N."] = datos_soja_filtered["DolarB.N."].astype(float)
datos_soja_filtered = datos_soja_filtered.dropna()
datos_soja_filtered["Anio"] = pd.to_datetime(datos_soja_filtered["Fecha"]).dt.year
#datos_soja_filtered.head(50)
```

```python
plt.figure(dpi=125)
ax = plt.gca()
ax.grid(axis='y', which='major')
ax.set_axisbelow(True)

sns.boxplot(data=datos_soja_filtered,y=datos_soja_filtered["Cierre"], palette=["#d7a75b"]) #Color de la semilla de soja

plt.ylabel("Precio por tonelada de soja(USD)")
plt.title("Fluctuaci√≥n del precio de soja a lo largo del tiempo")
plt.ylim(0,500)
plt.yticks(range(0, 501, 50))

plt.show()
```
